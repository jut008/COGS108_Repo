# Research Question

How much does home advantage and average attendance per game affect a team’s performance for the National Basketball Association vs. Major League Baseball? Are there any teams with a high win percentage on the road? Are there any teams with a low win percentage at home?

## Background and Prior Work


The topic of home court advantage has been discussed by fans and experts alike since the inception of modern organized sports. For certain professional sport leagues like the National Basketball Association and Major League Baseball, numerous studies have been conducted to investigate home court advantage and its impact on a team's performance. A recent study by Jason Kotecki dived deep into the psychology and reasoning behind why teams, specifically the National Basketball Association, performed better at home. Throughout Kotecki’s work, he explains other factors beyond the presence of the fans supporting their home team.  Kotecki explains that players do not want to shirk, or “purposely do not perform to the best of his ability”, in the presence of their fans ([^3](#cite_ref-3)).

Like the National Basketball Association, the Major League Baseball has also had studies conducted. According to Baseball Reference, historically the average of any team is .500, its winning percentage goes up to .530 at home, and down to .470 on the road ([^4](#cite_ref-4)). This ratio has held steady in Major League Baseball since 1945 and has been fairly consistent throughout all eras of baseball…” ([^5](#cite_ref-5)). While there have been some studies that have shown a correlation, we hope to confirm these findings, investigate any outliers within the data set, and determine what factors contributed to these anomalies.

## MERGED NBA DATA

```python
df_nba = pd.read_csv('NBA 2015-2019 Attendance.csv')
df_nba = df_nba.set_index(['YEAR', 'TEAM'])

for col in df_nba.columns:
    # Remove commas and convert to numeric
    df_nba[col] = df_nba[col].replace(',', '', regex=True)
    df_nba[col] = pd.to_numeric(df_nba[col], errors='coerce')

df_nba_wr = pd.read_csv('NBA W_R.csv')
df_nba_wr = df_nba_wr.set_index(['YEAR', 'TEAM']).rename(columns = {'HOME': 'HOME_WL', 'ROAD': "AWAY_WL"})
df_nba_wr = df_nba_wr[['HOME_WL', 'AWAY_WL']]

nba_data = pd.merge(df_nba, df_nba_wr, left_index=True, right_index=True, how='inner')
```

## MERGED MLB DATA

```python
df_mlb = pd.read_csv('MLB 2015-2019 Attendance.csv')
df_mlb = df_mlb.set_index(['YEAR', 'TEAM'])

for col in df_mlb.columns:
    # Remove commas and convert to numeric
    df_mlb[col] = df_mlb[col].replace(',', '', regex=True)
    df_mlb[col] = pd.to_numeric(df_mlb[col], errors='coerce')

df_mlb_winloss = pd.read_csv('MLB WinLoss.csv')
df_mlb_winloss = df_mlb_winloss.set_index(['YEAR', 'TEAM']).rename(columns = {'HOME': 'HOME_WL', 'AWAY': 'AWAY_WL'})
df_mlb_winloss = df_mlb_winloss[['HOME_WL', 'AWAY_WL']]

mlb_data = pd.merge(df_mlb, df_mlb_winloss, left_index=True, right_index=True, how='inner')
```

### Section 1 of EDA - Average Home vs. Away Win Percentages between Major League Baseball & National Basketball Association

In this section, we are displaying two bar graphs that showcase the Average Win Percentages at Home and Away for both Major League Baseball and National Basketball Association during the 2015 - 2019 seasons. Noted at the bottom, the data regarding Major League Baseball is on the left while the National Basketball Association’s data will be on the left. In our key, blue denotes Average Home Win Percentage while orange signifies Average Away Win Percentage.

```python
# Calculate average home and away win percentages for MLB and NBA
mlb_home_win_avg = mlb_data['HOME_WIN_PCT'].mean()
mlb_away_win_avg = mlb_data['AWAY_WIN_PCT'].mean()
nba_home_win_avg = nba_data['HOME_WIN_PCT'].mean()
nba_away_win_avg = nba_data['AWAY_WIN_PCT'].mean()

# Data for plotting
leagues = ['MLB', 'NBA']
home_win_avgs = [mlb_home_win_avg, nba_home_win_avg]
away_win_avgs = [mlb_away_win_avg, nba_away_win_avg]

# Creating the bar plot
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = np.arange(len(leagues))

bars1 = ax.bar(index, home_win_avgs, bar_width, label='Home Win %')
bars2 = ax.bar(index + bar_width, away_win_avgs, bar_width, label='Away Win %')

ax.set_xlabel('League')
ax.set_ylabel('Win Percentage')
ax.set_title('Average Win Percentage by Game Location for MLB and NBA')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(leagues)
ax.legend()

plt.tight_layout()
plt.show()
```

In this visualization, we opted for percentages rather than the number of games won at home and away to decrease any visual skew or bias. Because Major League Baseball has almost twice the amount of regular season games than the National Basketball association, it can cause the reader to assume based on visuals that Major League Baseball teams win more games at home. Using percentages, we deduced that overall, National Basketball Association teams on average perform better at home or worse away. It can also be said that Major League Baseball teams perform worse at home or better away.

### Section 2 of EDA - Distribution of Win Percentages Differences

We wanted to analyze each team's performance by checking if they won more home games or away games. To do so, we calculate the difference between "Home" and "Away" win-percentage for each team. This allows us to see if teams generally perform better at "Home" or "Away."

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load in data
mlb_data = pd.read_csv('MLB_DATA.csv')
nba_data = pd.read_csv('NBA_DATA.csv')

# Calculate the difference in win percentages for MLB and NBA
mlb_data['WIN_PCT_DIFF'] = mlb_data['HOME_WIN_PCT'] - mlb_data['AWAY_WIN_PCT']
nba_data['WIN_PCT_DIFF'] = nba_data['HOME_WIN_PCT'] - nba_data['AWAY_WIN_PCT']

# Plotting the distribution of win percentage differences for MLB
plt.figure(figsize=(14, 6))
sns.histplot(mlb_data['WIN_PCT_DIFF'], color='blue', kde=True, bins=30)
plt.title('MLB: Distribution of Difference in Win Percentages (Home - Away)')
plt.xlabel('Win Percentage Difference (Home - Away)')
plt.ylabel('Frequency')
plt.axvline(mlb_data['WIN_PCT_DIFF'].mean(), color='red', linestyle='--', label='Mean Difference')
plt.legend()
plt.show()

# Plotting the distribution of win percentage differences for NBA
plt.figure(figsize=(14, 6))
sns.histplot(nba_data['WIN_PCT_DIFF'], color='green', kde=True, bins=30)
plt.title('NBA: Distribution of Difference in Win Percentages (Home - Away)')
plt.xlabel('Win Percentage Difference (Home - Away)')
plt.ylabel('Frequency')
plt.axvline(nba_data['WIN_PCT_DIFF'].mean(), color='red', linestyle='--', label='Mean Difference')
plt.legend()
plt.show()
```

The plots above are showing the distribution of the difference in each team's win percentage. We can see it mostly follows a normal distribution. In the MLB distribution, we see it centers around +7.5%, indicating on average, teams win a higher percentage of their home games. In the NBA distribution, it centers around +16%, once again showing on average, teams win a higher percentage of their home games.

These analyses show that for both leagues, there's a general trend where teams win a higher percentage of their games when playing at "Home" vs. playing "Away". Indicating that a team is more likely to win a game if it is in their home court, but more so in the NBA.

### Section 4 of EDA - Win Percentage Difference for Each Team Per Year

The following graphs display the difference in win percentage (home win % - away win %) for each MLB and NBA team per year. The intention of these graphs is to show which teams perform better on the road than at home.

```python
mlb_2015 = mlb_data[mlb_data['YEAR'] == 2015]
mlb_2016 = mlb_data[mlb_data['YEAR'] == 2016]
mlb_2017 = mlb_data[mlb_data['YEAR'] == 2017]
mlb_2018 = mlb_data[mlb_data['YEAR'] == 2018]
mlb_2019 = mlb_data[mlb_data['YEAR'] == 2019]

fig, axes = plt.subplots(3, 2, figsize=(20, 18), constrained_layout=True)

# List of dataframes for iteration
dataframes = [mlb_2015, mlb_2016, mlb_2017, mlb_2018, mlb_2019]
titles = ['2015', '2016', '2017', '2018', '2019']

for i, (df, title) in enumerate(zip(dataframes, titles)):
    row = i // 2
    col = i % 2
    ax = axes[row, col]
    
    sns.barplot(x='TEAM', y=df['HOME_WIN_PCT'] - df['AWAY_WIN_PCT'], data=df, ax=ax)
    ax.set_title(f'Win Percentage Difference for each MLB Team in {title}')
    ax.set_xlabel('Team')
    ax.set_ylabel('Win Percentage difference')
    ax.tick_params(axis='x', rotation=90)

axes[-1, -1].axis('off')

plt.show()
```

The way we know if a team performed better on the road than at home is by looking to see if their bar is below 0 on the graph. If it is below 0, this means that the road win % was a bigger value than the home win % when taking the difference.
